```python
subcategory_map = mapping_dict = {
    'all': 'All',
    'price': 'Pricing',
    'emi': 'EMI & Loan',
    'location': 'Location',
    'features': 'Property features and amenities',
    'timeline': 'Timeline',
    'visit': 'Visit',
    'demography': 'Demographics'
}
```

```python
audio_df['Subcategory'] = audio_df['Subcategory'].map(subcategory_map)
```

```python
import re
def extract_and_create_filename(recording):
    match = re.search(r'callId=(\d+\.\d+)', recording)
    if match:
        call_id = match.group(1)
        return f"{call_id}.mp3"
    return recording

# Apply the function to the DataFrame
audio_df['Audio Name'] = audio_df['Audio Name'].apply(extract_and_create_filename)

audio_df.to_csv("per_audio1.csv", index=False)
```
